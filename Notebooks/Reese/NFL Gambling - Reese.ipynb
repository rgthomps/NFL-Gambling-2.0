```python
from sklearn.base import BaseEstimator
class GroupImputer(BaseEstimator):
    
    def __init__(self, fill_col, group_col, strategy):
        self.fill_col = fill_col
        self.group_col = group_col
        self.strategy = strategy
    
    def fit(self, X, y=None):
        self.val_by_group = X.groupby(self.group_col)[self.fill_col].agg([self.strategy]).to_dict()[self.strategy]
        
    def transform(self, X):
        X.loc[:, self.fill_col] = X[self.fill_col].fillna(X[self.group_col].map(self.val_by_group))
        return X
```

```python
group_imputer = GroupImputer(fill_col= 'weather_temperature', group_col='team_home', strategy='mean')
```

```python
class windImputer(BaseEstimator):
    
    def __init__(self, fill_col, wind_col, strategy):
        self.fill_col = fill_col
        self.wind_col = wind_col
        self.strategy = strategy
    
    def fit(self, X, y=None):
        self.val_by_wind = X.groupby(self.wind_col)[self.fill_col].agg([self.strategy]).to_dict()[self.strategy]
        
    def transform(self, X):
        X.loc[:, self.fill_col] = X[self.fill_col].fillna(X[self.wind_col].map(self.val_by_wind))
        return X
```

```python
wind_imputer = GroupImputer(fill_col= 'weather_wind_mph', group_col='team_home', strategy='mean')
```
